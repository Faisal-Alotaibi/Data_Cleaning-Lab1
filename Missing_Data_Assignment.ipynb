### Salary Column

#### Our approach is to get the average of positions with the same ID and to apply it to the missing value. And if there is still NaN after that we get the avg for the department. And for the CEO we set the max salary

```python
average_salary_by_position = df.groupby('PositionID')['Salary'].mean()
average_salary_by_position

average_salary_by_department = df.groupby('Department')['Salary'].mean()
average_salary_by_department
```

```python

df['Salary'] = df['Salary'].fillna(df['PositionID'].map(average_salary_by_position))
df['Salary'] = df['Salary'].fillna(df['Department'].map(average_salary_by_department))

ceo_salary = max(df['Salary'])
df.loc[df['PositionID'] == 16, 'Salary'] = ceo_salary
```

### DOB Column

### I Imputed by getting the DOB mode for each posID and if there is NaN after that we used the mode for the Department

```python
mode_dob_by_position = df.groupby('PositionID')['DOB'].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else pd.NaT)

df['DOB'] = df['DOB'].fillna(df['PositionID'].map(mode_dob_by_position))

mode_dob_by_department = df.groupby('Department')['DOB'].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else pd.NaT)

df['DOB'] = df['DOB'].fillna(df['Department'].map(mode_dob_by_department))

```
